### 돌린 이미지들 불러와서 csv로 만들기

```python
import glob
```

```python
# 0 - 29도 부터 -1도 부터 -30 도 까지
lists = []

for file in sorted(glob.glob('./Data/canvas2/*.jpg')):
    img = Image.open(file)
    lists.append(np.array(img).reshape(-1))
```

```python
df=pd.DataFrame(lists)
```
