```python
import glob
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
adidas_1 = sorted(glob.glob('./Images/Adidas_1/*.jpg')) # 0
adidas_2 = sorted(glob.glob('./Images/Adidas_2/*.jpg')) # 1
converse_1 = sorted(glob.glob('./Images/Converse_1/*.jpg')) # 2
converse_2 = sorted(glob.glob('./Images/Converse_2/*.jpg')) # 3
nike_1 = sorted(glob.glob('./Images/Nike_1/*.jpg')) # 4
nike_2 = sorted(glob.glob('./Images/Nike_2/*.jpg')) # 5
```

```python
train = np.zeros(1440 * 128 * 128 * 3).reshape(1440,128,128,3)
target = np.zeros(1440)
```

```python
i = 0
```

```python
for file in adidas_1:
    img = Image.open(file)
    if i >= 60 and i<= 179:
        img = img.transpose(Image.FLIP_LEFT_RIGHT)
        if i % 10 == 0:
            plt.imshow(img)
            plt.show()
    train[i] = np.array(img,dtype=np.int32)
    target[i] = 1
    i += 1
```

```python
for file in adidas_2:
    img = Image.open(file)
    train[i] = np.array(img,dtype=np.int32)
    target[i] = 2
    i += 1
```

```python
for file in converse_1:
    img = Image.open(file)
    train[i] = np.array(img,dtype=np.int32)
    target[i] = 3
    i += 1
```

```python
for file in converse_2:
    img = Image.open(file)
    train[i] = np.array(img,dtype=np.int32)
    target[i] = 4
    i += 1
```

```python
for file in nike_1:
    img = Image.open(file)
    train[i] = np.array(img,dtype=np.int32)
    target[i] = 5
    i += 1
```

```python
for file in nike_2:
    img = Image.open(file)
    train[i] = np.array(img,dtype=np.int32)
    target[i] = 6
    i += 1
```

```python
train_scaled = train.reshape(-1, 128 * 128 * 3) 
```

```python
import pandas as pd
```

```python
train_df =pd.DataFrame(train_scaled)
target_df = pd.DataFrame(target)
target_df.columns = ['target']
```

```python
df = pd.concat([train_df,target_df],axis=1)
```
