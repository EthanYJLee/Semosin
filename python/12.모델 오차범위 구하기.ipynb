```python
import keras
```

```python
model = keras.models.load_model('./Model/six-shoes-model.h5')
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./Data/train+target.csv')
df.head()
```

```python
import numpy as np

train = df.iloc[:,:-1]
train_arr = np.array(train)
```

```python
imgScaled = train_arr.reshape(-1,128,128,3)
```

```python
j = 0
predictList = []
for i in range(240 * j,240 * (j+1)):
     predictList.append(model.predict(imgScaled[i:i+1],verbose=0).tolist())
```

```python
# 오차범위 구하기
one = []
two = []
three = []
four = []
five = []

for pre in predictList:
    one.append(pre[0][1])
    two.append(pre[0][2])
    three.append(pre[0][3])
    four.append(pre[0][4])
    five.append(pre[0][5])
```

```python
oneList = [np.min(one),np.max(one)]
twoList = [np.min(two),np.max(two)]
threeList = [np.min(three),np.max(three)]
fourList = [np.min(four),np.max(four)]
fiveList = [np.min(five),np.max(five)]
```

```python
df = pd.DataFrame(
    [oneList,twoList,threeList,fourList,fiveList],
    index=['adidas2','converse1','converse2','nike1','nike2'],
    columns=['min','max']
)
```
